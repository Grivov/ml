```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

```python
X, y = mnist["data"], mnist["target"]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.imshow(X.iloc[0].values.reshape(28,28))
print(y[0])
```

```python
import numpy as np
y = y.astype(np.uint8)
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
y_train_5 = (y_train==5)
y_test_5 = (y_test==5)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
some_digit = X.iloc[[0]]  # Double brackets
prediction = sgd_clf.predict(some_digit)
print(prediction)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

```python
from sklearn.base import BaseEstimator
class Never5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        return self
    def predict(self, X):
        return np.zeros((len(X),1), dtype=bool)
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred_5 = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_train_pred_5, y_train_5)
conf
```

```python
precision = conf[1,1] / (conf[1,1]+conf[0,1])
recall = conf[1,1] / (conf[1,1]+conf[1,0])
f1_score = (1/precision/2 + 1/recall/2)**(-1)
print("precision = ", precision, " recall = ", recall, " f1 score = ", f1_score)
```
